```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r"C:\CO2Emission.xlsx")
df.head()
```

```python
#Tugas 6.1
#Perbandingan Kategori
#Perbandingan antara emisi GHG pada tahun 2002 dan 2012
df_yr_2012 = df.loc[df['Year'] == 2012]
df_yr_2002 = df.loc[df["Year"] == 2002]

df_TotalGas_exclude_2012 = df_yr_2012["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"].sum()
df_TotalGas_exclude_2002 = df_yr_2002["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"].sum()
df_TotalGas_include_2012 = df_yr_2012["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"].sum()
df_TotalGas_include_2002 = df_yr_2002["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"].sum()


judul1 = ["Excluding Land-Use Change and Forestry in 2012", "Excluding Land-Use Change and Forestry in 2002", "Including Land-Use Change and Forestry in 2012", "Including Land-Use Change and Forestry in 2002"]
isi1 = [df_TotalGas_exclude_2012, df_TotalGas_exclude_2002, df_TotalGas_include_2012, df_TotalGas_include_2002]

plt.barh(judul1 , isi1)
plt.title("The Total GHG in 2002 and 2012")
plt.xlabel("MtCO2e")
plt.grid()
plt.show()
```

```python
#Tugas 6.2
#Perbandingan Kategori
#Perbandingan emisi GHG pada 2012 antara Indonesia, Amerika Serikat, dan Cina
df_indonesia_2012 = df.loc[(df["Country"] == "Indonesia") & (df["Year"] == 2012)]
df_usa_2012 = df.loc[(df["Country"] == "United States") & (df["Year"] == 2012)]
df_cina_2012 = df.loc[(df["Country"] == "China") & (df["Year"] == 2012)]

df_TotalGas_indonesia_exclude_2012 = df_indonesia_2012["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"].sum()
df_TotalGas_indonesia_include_2012 = df_indonesia_2012["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"].sum()
df_TotalGas_usa_exclude_2012 = df_usa_2012["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"].sum()
df_TotalGas_usa_include_2012 = df_usa_2012["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"].sum()
df_TotalGas_cina_exclude_2012 = df_cina_2012["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"].sum()
df_TotalGas_cina_include_2012 = df_cina_2012["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"].sum()

judul2 = ["Indonesia Emission Excluding LUCF", "Indonesia Emission Including LUCF", "USA Emission Excluding LUCF", "USA Emission Including LUCF", "China Emission Excluding LUCF", "China Emission Including LUCF"]
isi2 = [df_TotalGas_indonesia_exclude_2012, df_TotalGas_indonesia_include_2012, df_TotalGas_usa_exclude_2012, df_TotalGas_usa_include_2012, df_TotalGas_cina_exclude_2012, df_TotalGas_cina_include_2012]

plt.barh(judul2, isi2)
plt.title("The Total of GHG in 2012 for China, USA, and Indonesia")
plt.xlabel("MtCO2e")
plt.grid()
plt.show()
```

```python
#Tugas 6.3
#Penampilan Perubahan Terhadap Waktu
# Jumlah emisi GHG pada tahun 1990-2012 
df_group_Year = df.groupby("Year")
df_year_totalghg_exclude = df_group_Year["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"].sum()
df_year_totalghg_include = df_group_Year["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"].sum()

df_year_totalghg_exclude.plot(kind = 'line', label = "Total GHG Excluding LUCF", color = "green")
df_year_totalghg_include.plot(kind = 'line', label = "Total GHG Including LUCF", color = "red")
plt.xlabel("Year")
plt.ylabel("MtCO2e")
plt.title("Total of GHG Emissions Including LUCF in 1990-2012")
plt.legend()
plt.show()
```

```python
#Tugas 6.4
#Penampilan Perubahan Terhadap Waktu
#Total emisi GHG pada 1990-2012 antara Indoneisa, Amerika Serikat, dan Cina 
df_indonesia = df.loc[df["Country"] == "Indonesia"]
df_cina = df.loc[df["Country"] == "China"]
df_usa = df.loc[df["Country"] == "United States"]

plt.plot(df_indonesia["Year"], df_indonesia["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"], label = "Indonesia Inc.", color = "red")
plt.plot(df_indonesia["Year"], df_indonesia["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"], linestyle = "--", label = "Indonesia Exc.", color = "red")

plt.plot(df_cina["Year"], df_cina["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"], label = "China", color = "orange")
plt.plot(df_cina["Year"], df_cina["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"], linestyle = "--", label = "China Exc.", color = "orange")

plt.plot(df_usa["Year"], df_usa["Total GHG Emissions Including Land-Use Change and Forestry (MtCO₂e‍)"], label = "USA", color = "blue")
plt.plot(df_usa["Year"], df_usa["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"], linestyle = "--", label = "USA Exc.", color = "blue")

plt.legend()
plt.show()
```

```python
#Tugas 6.5
#Penampilan Hierarki dan Hubungan Keseluruhan Bagian
#Emisi gas tanpa LUCF pada 2012
df_exc_co2_2012_1 = df_yr_2012["Total CO2 (excluding Land-Use Change and Forestry) (MtCO2)"].sum()
df_exc_ch4_2012_1 = df_yr_2012["Total CH4 (MtCO2e)"].sum()
df_exc_n2o_2012_1 = df_yr_2012["Total N2O (MtCO2e)"].sum()
df_exc_f_2012_1 = df_yr_2012["Total F-Gas (MtCO2e)"].sum()

isi3 = [df_exc_co2_2012_1, df_exc_ch4_2012_1, df_exc_n2o_2012_1, df_exc_f_2012_1]
judul3 = ["CO2", "CH4", "N2O", "F-Gas"]
plt.title("Emissions by Gas Excluding LUCF in 2012")
plt.pie(isi3, labels = judul3, wedgeprops = {'edgecolor' : 'black'}, shadow = True, autopct = "%1.1f%%", startangle = 15)
plt.show()
```

```python
#Tugas 6.7
#Penampilan Hierarki dan Hubungan Keseluruhan Bagian
#Emisi gas tanpa LUCF pada 2002
df_exc_co2_2002_1 = df_yr_2002["Total CO2 (excluding Land-Use Change and Forestry) (MtCO2)"].sum()
df_exc_ch4_2002_1 = df_yr_2002["Total CH4 (MtCO2e)"].sum()
df_exc_n2o_2002_1 = df_yr_2002["Total N2O (MtCO2e)"].sum()
df_exc_f_2002_1 = df_yr_2002["Total F-Gas (MtCO2e)"].sum()

isi5 = [df_exc_co2_2002_1, df_exc_ch4_2002_1, df_exc_n2o_2002_1, df_exc_f_2002_1]
judul5 = ["CO2", "CH4", "N2O", "F-Gas"]
plt.title("Emissions by Gas Excluding LUCF in 2002")
plt.pie(isi5, labels = judul5, wedgeprops = {'edgecolor' : 'black'}, shadow = True, autopct = "%1.1f%%", startangle = 15)
plt.show()
```

```python
#Tugas 6.9
#Penampilan Hierarki dan Hubungan Keseluruhan Bagian
#Sektor emisi pada 2012
df_energy = df_yr_2012["Energy (MtCO2e)"].sum()
df_industrial = df_yr_2012["Industrial Processes (MtCO2e)"].sum()
df_agriculture = df_yr_2012["Agriculture (MtCO2e)"].sum()
df_waste = df_yr_2012["Waste (MtCO2e)"].sum()
df_LUCF = df_yr_2012["Land-Use Change and Forestry (MtCO2)"].sum()
df_fuel = df_yr_2012["Bunker Fuels (MtCO2)"].sum()

isi7 = [df_energy, df_industrial, df_agriculture, df_waste, df_LUCF, df_fuel]
judul7 = ["Energy", "Industrial Processes", "Agriculture", "Waste", "LUCF", "Bunker Fuels"]
plt.title("Emissions by Sector in 2012")
plt.pie(isi7, labels = judul7, wedgeprops = {'edgecolor' : 'black'}, shadow = True, autopct = "%1.1f%%", startangle = 15)
plt.show()
```

```python
#Tugas 6.10
#Penampilan Hierarki dan Hubungan Keseluruhan Bagian
#Emisi energi pada sub sektor 2012 
df_electricity = df_yr_2012["Electricity/Heat (MtCO2)"].sum()
df_manufacturing = df_yr_2012["Manufacturing/Construction (MtCO2)"].sum()
df_transportation = df_yr_2012["Transportation (MtCO2)"].sum()
df_otherfuel = df_yr_2012["Other Fuel Combustion (MtCO2e)"].sum()
df_fugitive = df_yr_2012["Fugitive Emissions (MtCO2e)"].sum()

isi8 = [df_electricity, df_manufacturing, df_transportation, df_otherfuel, df_fugitive]
judul8 = ["Electricity", "Manufacturing", "Transportation", "Other Fuels", "Fugitive Fuels"]
plt.title("Energy Emissions by Sub-Sector in 2012")
plt.pie(isi8, labels = judul8, wedgeprops = {'edgecolor' : 'black'}, shadow = True, autopct = "%1.1f%%", startangle = 15)
plt.show()
```

```python
#Tugas 6.11
#Plotting Relationship
#Hubungan antara total emisi GHG terhadap energi pada 2012
df_total_ghg_exc_2012 = df_yr_2012["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"]
df_energy_2012 = df_yr_2012["Energy (MtCO2e)"]
df_waste_2012 = df_yr_2012["Waste (MtCO2e)"]

plt.scatter(df_total_ghg_exc_2012, df_energy_2012, c = df_waste_2012, edgecolor = "black", linewidth = 1, alpha = 0.65, s = 100)
cbar = plt.colorbar()
cbar.set_label("Waste (MtCO2e")
plt.xscale("log")
plt.yscale("log")

plt.title("Relationship Between Total GHG Emission and Energy in 2012 ")
plt.xlabel("Total GHG Emissions Excluding LUCF (MtCO2e)")
plt.ylabel("Energy (MtCO2e)")
plt.show()
```

```python
#Tugas 6.12
#Plotting Relationship
#Hubungan antara emisi total GHG terhadap listrik pada 2012
df_total_ghg_exc_2012 = df_yr_2012["Total GHG Emissions Excluding Land-Use Change and Forestry (MtCO2e)"]
df_electricity_2012 = df_yr_2012["Electricity/Heat (MtCO2)"]
df_waste_2012 = df_yr_2012["Waste (MtCO2e)"]

plt.scatter(df_total_ghg_exc_2012, df_electricity_2012, c = df_waste_2012, edgecolor = "black", linewidth = 1, alpha = 0.7, s = 100)
cbar = plt.colorbar()
cbar.set_label("Waste (MtCO2e")

plt.title("Relationship Between Total GHG Emission and Electricity in 2012 ")
plt.xlabel("Total GHG Emissions Excluding LUCF (MtCO2e)")
plt.ylabel("Electricity (MtCO2e)")
plt.show()
```
